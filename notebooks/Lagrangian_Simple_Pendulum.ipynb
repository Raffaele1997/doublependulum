```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

Lagrangian from Kinetic and Potential Energy of Simple Pendulum

```python
# Parameters
t, theta, m,g,l = symbols('t, theta, m,g,l')

# Time dependent Parameters
theta_t = Function('theta')(t)
theta_dot = diff(theta_t,t)
T  = 0.5*m*l**2*theta_dot**2

V = -m*g*l*cos(theta_t)

L = T-V

L
```

Euler Lagrange Equations

```python
# Linker Term: d/dt ( dL / d(theta_dot) )
dL_dtheta_dot = diff(L, theta_dot)
d_dt_dL_dtheta_dot = diff(dL_dtheta_dot, t)

# Rechter Term: dL / d(theta)
dL_dtheta = diff(L, theta_t)

# 4. Die Gleichung aufstellen: d/dt(...) - dL/dtheta = 0
euler_lagrange_eq = Eq(d_dt_dL_dtheta_dot - dL_dtheta, 0)

# Anzeige der Gleichung
display(euler_lagrange_eq)
```

```python
# Nach theta_dot_dot auflösen
theta_ddot = diff(theta_t, (t, 2))
accel_expr = solve(euler_lagrange_eq, theta_ddot)[0]
# Umwandlung in eine numerische Funktion (Lambdify)
import numpy as np
from scipy.integrate import solve_ivp
from sympy import lambdify

# Wir erstellen eine Funktion für die Beschleunigung: f(g, l, theta)
func_accel = lambdify((g, l, theta_t), accel_expr, 'numpy')
func_accel
```
