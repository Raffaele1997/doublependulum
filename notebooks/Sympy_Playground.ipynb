# Importing Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

Make a a function $$f(x) = \sin{ax + bx + c}$$

```python
x,a,b,c = symbols('x a b c')

function = sin(a*x**2 + b*x + c)

function
```

Differentiate $f(x)$

```python
function_diff = diff(function, x)

function_diff
```
